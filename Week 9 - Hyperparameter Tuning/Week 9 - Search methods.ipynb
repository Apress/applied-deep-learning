# Search of Maximum

## Grid search vs. Random search

Let's first define a function with a maximum. Goal is to find the maximum as best as possible with the less number of tries possible.

````python
def maxim(x):
    tmp1 = -np.cos(x/4.0)-np.sin(x/4.0)-2.5*np.cos(2.0*x/4.0)+0.5*np.sin(2.0*x/4.0)
    tmp2 = -np.cos(x/3.0)-np.sin(x/3.0)-2.5*np.cos(2.0*x/3.0)+0.5*np.sin(2.0*x/3.0)
    return 10.0+tmp1+0.5*tmp2
````

````python
x = np.arange(0, 80, 0.1)
````

````python
plt.plot(x, maxim(x))
plt.show()
````

````python
np.random.seed(7)
````

# Bayesian Optimisation

````python
def K(x, l, sigm = 1):
    return sigm*np.exp(-1.0/2.0/l**2*(x**2))
````

Let's choose first some random points from a function f

````python
def f(x):
    return x**2-x**3+3+10*x+0.07*x**4
````

````python
x_ = np.arange(0,12,0.1)
y_ = f(x_)
````

````python
plt.rc('font', family='arial')
plt.rc('xtick', labelsize='x-small')
plt.rc('ytick', labelsize='x-small')
    
plt.tight_layout()

fig = plt.figure(figsize=(8,4))

plt.plot(x_,y_, color = 'black')
plt.xlabel('$x$', fontsize = 16)
plt.ylabel('$f(x)$', fontsize = 16)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)

plt.show()
````

Let's choose 5 random points

````python
np.random.seed(42)
randompoints = np.random.random([5])*12.0
````

````python
f_ = f(randompoints)
````

````python

plt.rc('font', family='arial')
plt.rc('xtick', labelsize='x-small')
plt.rc('ytick', labelsize='x-small')
    
plt.tight_layout()

fig = plt.figure(figsize=(8,4))

plt.plot(x_,y_, color = 'black')
plt.scatter(randompoints, f_, marker = 'x', s = 70, color = 'black')
plt.xlabel('$x$', fontsize = 16)
plt.ylabel('$f(x)$', fontsize = 16)

plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)

plt.show()
````

let's build the necessary components

````python
xsampling = np.arange(0,14,0.2)

np.random.seed(42)
randompoints = np.random.random([5])*12.0

ybayes_ = []
sigmabayes_ = []
for x in xsampling:
    
    
    f1 = f(randompoints)
    sigm_ = np.std(f1)**2
    f_ = (f1-np.average(f1))
    
    
    k = K(x-randompoints, 2 ,sigm_)
    
    C = np.zeros([randompoints.shape[0], randompoints.shape[0]])
    Ctilde = np.zeros([randompoints.shape[0]+1, randompoints.shape[0]+1])
    for i1,x1_ in np.ndenumerate(randompoints):
        for i2,x2_ in np.ndenumerate(randompoints):
            C[i1,i2] = K(x1_-x2_, 2.0,sigm_)
            
    Ctilde[0,0] = K(0, 2.0,sigm_)
    Ctilde[0,1:randompoints.shape[0]+1] = k.T
    Ctilde[1:,1:] = C
    Ctilde[1:randompoints.shape[0]+1,0] = k
    
    mu = np.dot(np.dot(np.transpose(k), np.linalg.inv(C)), f_) 
    sigma2 = K(0, 2.0,sigm_)- np.dot(np.dot(np.transpose(k), np.linalg.inv(C)), k)
    ybayes_.append(mu)
    sigmabayes_.append(np.abs(sigma2))
    
ybayes = np.asarray(ybayes_)+np.average(f1)
sigmabayes = np.asarray(sigmabayes_)
````

````python
plt.rc('font', family='arial')
plt.rc('xtick', labelsize='x-small')
plt.rc('ytick', labelsize='x-small')

fig = plt.figure(figsize=(10,4))

plt.plot(x_,y_, color = 'black', label = 'original function')
plt.scatter(randompoints, f1, marker = 'x', s = 70, color = 'black', label = 'randomly selected points')
plt.plot(xsampling, ybayes, color = 'black', ls = '--', label = 'Predicted function')
plt.fill_between(xsampling, ybayes-np.sqrt(sigmabayes), ybayes+np.sqrt(sigmabayes), alpha = 0.3, color = 'black')

plt.xlabel('$x$', fontsize = 16)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., fontsize = 16)

plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)

plt.show()
````

# Function form of Bayesian method

````python
def get_surrogate(randompoints):
    ybayes_ = []
    sigmabayes_ = []
    for x in xsampling:
    
        f1 = maxim(randompoints)
        sigm_ = np.std(f1)**2
        f_ = (f1-np.average(f1))
        k = K(x-randompoints, 2.0, sigm_ )

        C = np.zeros([randompoints.shape[0], randompoints.shape[0]])
        Ctilde = np.zeros([randompoints.shape[0]+1, randompoints.shape[0]+1])
        for i1,x1_ in np.ndenumerate(randompoints):
            for i2,x2_ in np.ndenumerate(randompoints):
                C[i1,i2] = K(x1_-x2_, 2.0, sigm_)

        Ctilde[0,0] = K(0, 2.0)
        Ctilde[0,1:randompoints.shape[0]+1] = k.T
        Ctilde[1:,1:] = C
        Ctilde[1:randompoints.shape[0]+1,0] = k

        mu = np.dot(np.dot(np.transpose(k), np.linalg.inv(C)), f_)
        sigma2 = K(0, 2.0, sigm_)- np.dot(np.dot(np.transpose(k), np.linalg.inv(C)), k)
        ybayes_.append(mu)
        sigmabayes_.append(np.abs(sigma2))

    ybayes = np.asarray(ybayes_)+np.average(f1)
    sigmabayes = np.asarray(sigmabayes_)

    return ybayes, sigmabayes
````

````python
def get_new_point(ybayes, sigmabayes, aa):
    newpoint = np.max(np.average(ybayes)+aa*np.sqrt(sigmabayes))

    idxmax = np.argmax(np.average(ybayes)+aa*np.sqrt(sigmabayes))
    newpoint = xsampling[idxmax]
    
    return newpoint
````

````python
def get_new_point2(ybayes, sigmabayes, aa):
    newpoint = np.max(ybayes+aa*np.sqrt(sigmabayes))

    idxmax = np.argmax(ybayes+aa*np.sqrt(sigmabayes))
    newpoint = xsampling[idxmax]
    
    return newpoint
````
